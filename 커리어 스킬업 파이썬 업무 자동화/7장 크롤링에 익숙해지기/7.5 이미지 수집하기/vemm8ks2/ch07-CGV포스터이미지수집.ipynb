```python
import requests
from bs4 import BeautifulSoup

url = 'http://www.cgv.co.kr/movies/?lt=1&ft=0'

code = requests.get(url)
html = BeautifulSoup(code.text, 'html.parser')

title_elements = html.select('strong.title')
image_elements = html.select('span.thumb-image > img')

for title, image in zip(title_elements, image_elements):
    print(f"제목 : {title.text}")
    print(image.attrs["src"])
    print("------------------\n")
```

```python
import urllib.request as req
import os

folder_name = "C:/CGV_무비차트_포스터"

if not os.path.exists(folder_name):
    os.mkdir(folder_name)

num = 1

for title, image in zip(title_elements, image_elements):
    print(f"제목 : {title.text}")
    img_url = image.attrs['src']
    req.urlretrieve(img_url, f'{folder_name}/{num}.jpg')
    num += 1
    print()
```
