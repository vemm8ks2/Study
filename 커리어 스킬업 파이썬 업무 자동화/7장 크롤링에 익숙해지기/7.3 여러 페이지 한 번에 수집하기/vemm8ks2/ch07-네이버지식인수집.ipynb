```python
import requests
from bs4 import BeautifulSoup

keyword = input('키워드 입력:')
print()
url = f'https://kin.naver.com/search/list.naver?query={keyword}'

code = requests.get(url)
html = BeautifulSoup(code.text, 'html.parser')

title = html.select('ul.basic1 a._nclicks\\:kin\\.txt._searchListTitleAnchor')
date = html.select('dd.txt_inline')

for t, d in zip(title, date):
    print(f'질문 : {t.text}')
    print(f'날짜 : {d.text}\n')
```

```python
for page in range(1, 11):
    code = requests.get(f'{url}&page={page}')
    html = BeautifulSoup(code.text, 'html.parser')

    title = html.select('ul.basic1 a._nclicks\\:kin\\.txt._searchListTitleAnchor')
    date = html.select('dd.txt_inline')
    
    print(f'----- {page}페이지 시작 -----\n')

    for t, d in zip(title, date):
        print(f'질문 : {t.text}')
        print(f'날짜 : {d.text}\n')

    print(f'----- {page}페이지 종료 -----\n\n')
```
