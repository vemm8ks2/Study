### 한 페이지만 수집하기

````python
import requests
from bs4 import BeautifulSoup

keyword = input("키워드를 입력해주세요 : ")
code = requests.get(f"https://kin.naver.com/search/list.naver?query={keyword}")
soup = BeautifulSoup(code.text,"html.parser")
title = soup.select("ul.basic1 a._nclicks\\:kin\\.txt._searchListTitleAnchor")
date = soup.select("dd.txt_inline")
for i,j in zip(title, date):
    print(f"질문 : {i.text}")
    print(f"날짜 : {j.text}")
    print("-------------------------------------------------")
````

### 여러 페이지 수집하기

````python
import requests
from bs4 import BeautifulSoup

keyword = input("키워드를 입력해주세요 : ")
for page_num in range(1,11):
    code = requests.get(f"https://kin.naver.com/search/list.naver?query={keyword}&page={page_num}")
    soup = BeautifulSoup(code.text, "html.parser")
    title = soup.select("ul.basic1 a._nclicks\\:kin\\.txt._searchListTitleAnchor")
    date = soup.select("dd.txt_inline")
    for i,j in zip(title,date):
        print(f"질문 : {i.text}")
        print(f"날짜 : {j.text}")
        print("-------------------------------------------------")
    
````

#### 3. 다음 요소를 지정하는 CSS 선택자로 적절한 것을 고르세요.
```
< span class = "ab:cd ef.gh" >
```
#### 1."span.ab:cd ef.gh"
#### 2."span.ab\:cd ef\.gh"
#### 3."span.ab\:cd\ ef\.gh"
#### 4."span.ab\:cd.ef.gh"
##### 정답 : 2
##### 해설 : (:)콜론과 (.)점 앞에는 \(이스케이프 코드를 )넣고 공백을 마침표로 작성해야 문자열로 인식하기 때문에 정답은 "span.ab\:cd ef\.gh"인 2번 입니다.
