```python
import requests
from bs4 import BeautifulSoup

url = f"https://underkg.co.kr/news"

code = requests.get(url)
html = BeautifulSoup(code.text, 'html.parser')

title = html.select('h1.title > a')

for i in title:
    print(f"제목 : {i.text}")
    print(i.attrs['href'], '\n')
```

```python
for i in title:
    print(f"제목 : {i.text}")
    
    news_url = i.attrs['href']
    print(f"링크 : {news_url}")
    
    code_news = requests.get(news_url)
    html_news = BeautifulSoup(code_news.text, 'html.parser')
    
    content = html_news.select_one('div.read_body')
    print(content.text.replace('\n', '').strip())
    print('\n--------------------------------------------------\n')
```
